```
import pandas as pd
```

```
df = pd.read_csv("https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/iris.csv")
df
```

```
df.groupby('Name').count()
```

#Parallel coordinates

```
pd.plotting.parallel_coordinates(df,'Name')
```

```
pd.plotting.parallel_coordinates(df,'Name',color = ('#4c75a1','#9740bd','#2736db')) #กำหนดสีด้วย html colots
```

```
pd.plotting.parallel_coordinates(df,'Name',color = ('r','g','b')) #กำหนดสีด้วยตัวย่อสี R G B W C M Y K
```

```
pd.plotting.parallel_coordinates(df,'Name',color = ((1,0,0),(0,1,0),(0,0,1))) #กำหนดสีด้วย rgb value (r,g,b) เป็นการกำหนดค่าอสงที่มาผสมกัน r,g,b -> 0-1
```

```
pd.plotting.parallel_coordinates(df,'Name',color = ((0.8,0.2,0.9),(0.6,0.5,0.3),(0,0,0))) #ปรับได้สิบหกล้านสี
```

##กำหนด paramiter colormaps 

```
pd.plotting.parallel_coordinates(df,'Name',colormap = 'cool') #colormap เป็นสีที่เขากำหนดชุดสีไว้อยู่แล้ว
```
